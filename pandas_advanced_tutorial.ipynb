```python
import pandas as pd
data = pd.read_csv("file.csv")
```

```python
data.columns
```

```python
#drop a column in dataframe
#here axis=0 represents index droping , axis =1 represents column dropping
data = data.drop(['Unnamed: 0'],axis=1)
len(data.columns)
```

```python
#checking the info about the data
data.info()
data.isnull().sum()
```

```python
# major operations in pandas
#print(data.Product_Color.unique())
# to print total number of observations in dataset
data.shape
data.head()
data['order_date'] = pd.to_datetime(data['order_date'])
data['year'] = pd.DatetimeIndex(data['order_date']).year
data['month'] = pd.DatetimeIndex(data['order_date']).month
#pd.DatetimeIndex is a subclass of metadata that carry metadata such as frequency of information 
```

```python
#indeexing of operations 
#loc[],iloc[],ix[],[]
#data.head()
#first== reterving only particular columns
data[['Name','Price','Product_Category']]
# second == selecting only single row = .loc(selects particular column based on the labels in rows in columns)
data.loc[0:9] #selects based on the index - it reterives 8 rows 
first = data.loc[[0,9], 
                   ["Name",'year']] 
second = data.loc[:, 
                   ["Name",'year']] # selects all the rows and specified columns
# third by position == iloc[6]-reterives sixth position rows
data.iloc[[7],[4]] # by indexinteger position [rows and columns]
data.iloc[[0,5],[0,7,6]]
data.iloc[[19],[7]]

#fourth by label and integer based ,however it is deprecated in recent versions of pandas, for example
data.ix[0]
data.iloc[[19],[7]] #19 th column product_name
```

```python
# summarizing operations 
data.describe()# summerizes only integer columns
data.describe(include="all")# it includes all the columns
data.Name.describe() # summerizes only one column 
# most and least occurences
data.Product_Category.value_counts().head()#lists number of times the product is used
data.Product_Category.value_counts().tail()# least occurences
```

```python
# Filtering and sorting in pandas
# six ways to sort in pandas
# first === sorting values by column 
col_sort = data.sort_values('Product_Category')# sorts by product_category by first asc and last by desc
# second ===  sort df based on values of column by desc order
col_desc_sort = data.sort_values('Price',ascending= False)
#print(col_desc_sort)
# third === sort values and put missing values first in the df 
col_missing_value_first = data.sort_values('Price',na_position='first')
#print(col_missing_value_first)
# fourth === sort dataframe based on column in place,i.e without creating new dataframe 
col_inplace_sort = data.sort_values('Price',inplace = True,ascending = False)
#print(col_inplace_sort)
#fifth = sort df based on index 
index_sorting = data.sort_index(inplace=True)
#print(index_sorting)
# sort values by multiple columns
multiple_sort = data.sort_values(['Name','Price'])
multiple_sort
```

```python
import numpy as np
# grouping operations
# works on split-apply-combiine method
#spliting the groups based on certain condition
# appling function to all the groups
# combaining into final data structure
#first == based on product category the average product range is taken 
data_split_grouping = data.groupby(['Product_Category']).Price.count()
# second == based on multiple columns grouping 
data_muliple_split_grouping = data.groupby(['Product_Category','Name']).Total_Computed.sum().reset_index()# reset_index resets the index to original position
data_muliple_split_grouping
# third == applying multiple functions at once  without using numpy and lambda method
data_apply_func = data.groupby(['Product_Category']).agg({'Price':['count','mean'], 
                         'Total_Computed':'sum'}).reset_index()
# fourth == using np  applying multiple functions 
data_apply_numpy = data.groupby(['Product_Category']).agg([np.sum, np.mean, np.std])
# fifth  == grouping through particular category
data_lighting_grouping = data[data['Product_Category'] == 'Lighting'].groupby('Product_Category').Price.mean()
data_lighting_grouping
# for reference https://github.com/guipsamora/pandas_exercises/blob/master/03_Grouping/Occupation/Exercises_with_solutions.ipynb
```
